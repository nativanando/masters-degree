# Iris dataset!

```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
df = pd.read_csv(url, names = ['sepala_comprimento','sepala_largura','petala_comprimento','petala_largura','classe'])
print (df.describe())
```

```python
print('Moda comprimento da sepala:', df['sepala_comprimento'].mode())
print('Moda largura da sepala:', df['sepala_largura'].mode())
print('Moda comprimento da petala:', df['petala_comprimento'].mode())
print('Moda largura da petala:', df['petala_largura'].mode())
```

```python
print(df.corr())
```
